```python
# Group the AGE in bins of 10 years
labels = ["{0} - {1}".format(i, i + 9) for i in range(1, 70, 10)]

kpmg['age_group'] = pd.cut(kpmg.AGE, range(1, 80, 10), right=False, labels=labels)
```
